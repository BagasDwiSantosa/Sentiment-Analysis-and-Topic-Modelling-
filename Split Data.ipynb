```python
def categorize_rating(rating):
    if rating >= 4:
        return 'positif'
    elif rating == 3:
        return 'netral'
    elif rating <=2:
        return 'negatif'

# Tambahkan kolom 'Label' berdasarkan rating menggunakan fungsi categorize_rating
df['label'] = df['rating'].apply(categorize_rating)

df
```
