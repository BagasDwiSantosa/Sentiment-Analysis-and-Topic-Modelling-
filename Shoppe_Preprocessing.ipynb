### Data Selection

```python
def remove_duplicate_words(text):
    words = text.split()  # Split text into words
    unique_words = []
    for word in words:
        if word not in unique_words:
            unique_words.append(word)
    cleaned_text = ' '.join(unique_words)  # Join the unique words back into text
    return cleaned_text

df['comment2'] = df['comment2'].apply(remove_duplicate_words)
df
```

### Stopword Removal

```python
!pip install Sastrawi
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Initialize Sastrawi components
stopword_remover = StopWordRemoverFactory().create_stop_word_remover()
stemmer = StemmerFactory().create_stemmer()

# Function to remove unclear words
def remove_unclear_words(text):
    # Remove stopwords
    text_without_stopwords = stopword_remover.remove(text)

    # Tokenize the text into words
    words_in_text = text_without_stopwords.split()

    # Stemming the words
    stemmed_words = [stemmer.stem(word) for word in words_in_text]

    # Join the stemmed words back into a cleaned text
    cleaned_text = ' '.join(stemmed_words)
    return cleaned_text

# Clean and preprocess the example text
df['Stopword'] = df['comment2'].swifter.apply(remove_unclear_words)
df
```

### Stemming

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Initialize Sastrawi stemmer
stemmer_factory = StemmerFactory()
stemmer = stemmer_factory.create_stemmer()

# Function to perform lemmatization using Sastrawi
def lemmatize_indonesian_text(text):
    words = text.split()  # Split text into words
    lemmatized_words = [stemmer.stem(word) for word in words]  # Perform stemming
    lemmatized_text = ' '.join(lemmatized_words)  # Join the lemmatized words back into text
    return lemmatized_text

df['Stemming'] = df['Stopword'].swifter.apply(lemmatize_indonesian_text)
df
```
